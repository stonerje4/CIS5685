```python
import pandas as pd

df = pd.read_csv("boone_clean.csv")

df.head
```

```python
df.describe()
```

```python
df = df.groupby(pd.cut(df.index, 4)).mean()
df = df.drop(columns=["Unnamed: 0","WeekIndex","StoreNum"])
df = df.reset_index(drop=True)
df["Quarter"] = ["Q1","Q2","Q3","Q4"]   
df.round(2)
```

```python
#rearrange the columns 
quarter = df.pop('Quarter')
df.insert(0, 'Quarter', quarter)
df.round(2)
```

```python
df.to_csv("boone_summary.csv", index=False)
```
